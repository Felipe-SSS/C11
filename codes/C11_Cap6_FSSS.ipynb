> <h4 style="font-family:Lucida Console">Exercício 1</h4>

````python
import pandas as pd
import matplotlib.pyplot as plt

dfPaises = pd.read_csv('../datasets/paises.csv', delimiter=';')
````

````python
df_na = dfPaises[dfPaises["Region"].str.contains("NORTHERN AMERICA", na=False)]

df_na = df_na.sort_values("Country")

plt.figure(figsize=(10,6))
````

````python
# Taxa de mortalidade
plt.plot(df_na["Country"], df_na["Deathrate"], label="Taxa de Mortalidade", marker="o")
# Taxa de natalidade
plt.plot(df_na["Country"], df_na["Birthrate"], label="Taxa de Natalidade", marker="s")

plt.title("Taxa de Mortalidade e Natalidade - América do Norte")
plt.xlabel("País")
plt.ylabel("Taxa (%)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
````

> <h4 style="font-family:Lucida Console">Exercício 2</h4>

````python
import pandas as pd
import matplotlib.pyplot as plt

dfSpace = pd.read_csv('../datasets/space.csv', delimiter=';')
````

````python
df_usa = dfSpace[dfSpace["Location"].str.contains("USA", na=False)]
df_china = dfSpace[dfSpace["Location"].str.contains("China", na=False)]

num_empresas_usa = df_usa["Company Name"].nunique()
num_empresas_china = df_china["Company Name"].nunique()

# Separa os dados dos países para ajudar a plotar
dados = {
    "USA": num_empresas_usa,
    "China": num_empresas_china
}
````

````python
plt.figure(figsize=(6,4))
plt.bar(dados.keys(), dados.values(), color=["blue", "red"])

plt.title("Número de Empresas Espaciais por País")
plt.ylabel("Quantidade de Empresas Únicas")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
````

> <h4 style="font-family:Lucida Console">Exercício 3</h4>

````python
import pandas as pd
import matplotlib.pyplot as plt

dfSpace = pd.read_csv('../datasets/space.csv', delimiter=';')
````

````python
roscosmos = dfSpace[dfSpace["Company Name"] == "Roscosmos"]

contagem = roscosmos["Status Mission"].value_counts()

contagem = contagem[contagem.index.isin(["Success", "Failure"])]
````

````python
plt.figure(figsize=(6,6))
plt.pie(contagem, labels=contagem.index, autopct="%1.1f%%", colors=["green", "red"], startangle=90)

plt.title("Missões da Roscosmos: Sucesso vs Falha")

plt.show()
````
